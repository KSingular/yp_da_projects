### Обработка дубликатов в названиях объектов общественного питания

Изучим столбец с названиями объектов общественного питания, проверим его содержимое на ошибки и скрытые дубликаты (разное написание одного и того же названия).  

Продублируем столбец с названиями, чтобы сохранить исходные данные: 

```python
rest_data['object_name_old'] = rest_data['object_name']
```

Узнаем сколько датасет содержит уникальных названий объектов:

```python
print('Количество уникальных названий объектов общепита:', rest_data['object_name'].nunique())
```

Удалим кавычки и вновь посчитаем уникальные названия:

Проверим датасет на дубликаты, для чего отберем столбцы по названию объекта, адресу, типу объекта и количеству посадочных мест:

```python
print('Количество дубликатов:', len(rest_data[rest_data\
      .duplicated(subset=['object_name', 'address', 'object_type', 'number']) == True]))
```

Удалим обнаруженные дубликаты:

```python
lenrestdata = len(rest_data) # Переменная для вычисления процента очистки датасета
rest_data = rest_data.drop_duplicates(subset=['object_name', 'address', 'object_type', 'number'])\
                     .reset_index(drop=True)
```

### Определение принадлежности к сети заведений общественного питания

Для определения принадлежности объекта общественного питания к сети применим алгоритм редакторского расстояния Левенштейна, который поможет найти названия схожие с заданными ключами: 

```python
# Функция "Алгоритм преобразования Левенштейна"
def leven(data, key_list, sim):
    actual_list = []
    similarity = []
    for i in data.object_name:
        ratio = process.extractOne(i, key_list, scorer=fuzz.token_set_ratio)
        actual_list.append(ratio[0])
        similarity.append(ratio[1]) 
    
    data['actual_list'] = pd.Series(actual_list)
    data['similarity'] = pd.Series(similarity)     
    data.loc[(data.similarity < sim), ('actual_list', 'similarity')] = np.nan
    # data.style.background_gradient(subset='similarity', cmap='coolwarm')
```

Для лучшей работы алгоритма удалим из названий объектов общественного питания префексы (типы объектов) с помощью регулярных выражений:

```python
print('Количество уникальных названий объектов:', rest_data['object_name'].nunique())
```

### Исследование количества посадочных мест для сетевых заведений общественного питания

Исследуем количество посадочных мест для сетевых заведений, для чего создадим сводную таблицу с расчетом количества заведений входящих в сеть и средним количеством посадочных мест, построим график:

```python
catering_piv = catering_data[catering_data['chain_name'].isna() == False]\
    .pivot_table(values='number', index='chain_name', aggfunc=['count', 'mean']).reset_index(drop=False)
catering_piv['mean'] = catering_piv['mean'].astype(int)
catering_piv.columns = ['chain_name', 'ncount', 'nmean']
catering_piv[0:2]
```

```python
forx=catering_piv.quantile(.75)[0]
fory=catering_piv.quantile(.75)[1]
fig_number = plt.figure(figsize=(12, 4))
ax_number = sns.scatterplot(data=catering_piv, x='ncount', y='nmean', color='#FF4500', alpha=0.6)
ax_number.margins(0.02, 0.05)
ax_number.set_title('Отношение между количеством заведений в сети и количеством посадочных мест',
                    fontweight='bold', pad=18)
ax_number.set_xlabel('Количество заведений в сети общественного питания', fontweight='bold')
ax_number.set_ylabel('Среднее количество посадочных мест', fontweight='bold')
ax_number.set_xticks(np.arange(0, 185, 10))
ax_number.text(151, 265, 'rs = {:.2f}'.format(catering_piv['nmean'].corr(catering_piv['ncount'],
                                              method='spearman')), fontweight='bold', fontsize=14)
ax_number.axvline(x=forx, c='grey', linestyle ='dashed')
transx = transforms.blended_transform_factory(ax_number.transData, ax_number.get_xticklabels()[0].get_transform())
ax_number.text(forx, 0, "{:.0f}".format(forx), color='red', transform=transx, ha='right', va='top')
ax_number.axhline(y=fory, c='grey', linestyle ='dashed')
transy = transforms.blended_transform_factory(ax_number.get_yticklabels()[0].get_transform(), 
                                             ax_number.transData)
ax_number.text(0, fory, "{:.0f}".format(fory), color='red', transform=transy, ha='right', va='center')
plt.show()
```

На графике видно, что в большинство сетей общественного питания входит до 8 заведений со средним количеством посадочных мест до 76 (на уровне 3 квартиля). Отмечена некоторая тенденция к тому, что чем больше заведений в сети, тем меньше посадочных мест и наоборот, однако корреляционный анализ ее не подтверждает (rs = -0,07).    
  
Таким образом, размер сети и количество посадочных мест являются независимыми показателями.

### Исследование зависимости количества посадочных мест от типа объекта общественного питания

Исследуем зависимость количества посадочных мест от типа объекта общественного питания, для чего построим график:

```python
ord_s = catering_data.groupby('object_type')['number'].median().sort_values(ascending=False).index
fig_type_number = plt.figure(figsize=(12, 6))
ax_type_number = sns.boxplot(data=catering_data, x='number', y='object_type', order=ord_s,
                             palette='Blues_r', showmeans=False, showfliers=False)
ax_type_number.margins(0.02,0)
ax_type_number.set_title('График зависимости количества посадочных мест от типа \
объекта общественного питания', fontweight='bold', pad=18)
ax_type_number.set_xlabel('Количество посадочных мест', fontweight='bold')
ax_type_number.set_ylabel('Тип объекта общественного питания', fontweight='bold')
medians = catering_data.groupby('object_type').number.agg(['count', 'median'])\
                       .sort_values('median', ascending=False)['median'].values
median_labels = [str(int(s)) for s in medians]
pos = range(len(medians))
for tick, label in zip(pos, ax_type_number.get_yticklabels()):
    ax_type_number.text(medians[tick], pos[tick]+0.05, median_labels[tick], 
                        size=8, color='white', weight='semibold', bbox=dict(facecolor='#CD5C5C'))
plt.show()
```

Наибольшее количество посадочных мест характерно для столовой - медианное значение 103 места. Для ресторана эта цифра составляет 80 мест, для бара - 35, для кафе и буфета - 32. ### Выделение в отдельный столбец информации об улицах, на которых расположены заведения

Выделим в отдельный столбец информацию об улицах, на которых расположены заведения. В адресах присутствует префикс "город Москва", удалим его и скорректируем некоторые адреса:

```python
# Сохранение столбца с исходныи данными
merged_data['address_old'] = merged_data['address']
# Удаление префексов и замена неверных данных
merged_data['address'] = merged_data['address'].str.replace(r'город Москва, ','', regex = True)
merged_data['address'] = merged_data['address'].str.replace(r'ё','e', regex=True)
merged_data['address'] = merged_data['address'].str.replace(r'улица Мичуринский Проспект, Олимпийская Деревня',
                                                             'улица Мичуринский Проспект Олимпийская Деревня', regex = True)
```

Для определения московских адресов напишем и запустим функцию:

```python
# Функция для определения московских адресов
merged_data['address_1'] = merged_data['address'].str.split(', ', expand=True)[0]
def street_moscow(data):
    if data['address_1'].startswith(tuple(['город', 'поселение'])) == False:
        return data['address_1']
    return np.nan
merged_data['street_moscow'] = merged_data.apply(street_moscow, axis = 1)
```

Также напишем и запустим функцию для определения "немосковских адресов". Фактически данные города и поселения являются новыми районами Москвы. Улицы данных населенных пунктов могут совпадать и вводить в заблуждение, поэтому для них сохраним префексы:

```python
def street_nomoscow(data):
    if data['address_1'].startswith(tuple(['город', 'поселение'])) == True:
        return str(data['address_1']) + ', ' + str(data['address_2'])
    return data['address_1']
merged_data['address_1'] = merged_data['address'].str.split(', ', expand=True)[0]
merged_data['address_2'] = merged_data['address'].str.split(', ', expand=True)[1]
merged_data['street_nomoscow'] = merged_data.apply(street_nomoscow, axis = 1)
```

В первую очеред стоит исследовать следующие районы - Пресненский, Тверской, Басманный, Даниловский, Замоскворечье, Хамовники, Таганский, Мещанский и Арбат.  
4. Подбор помещения стоит начать в первую очередь на улицах с наименьшим количеством конкурентов, однако стоит учитывать, что отсутствие заведений общественного питания может быть связано с низкой проходимостью локации - отсутствием офисов, жилых домов, туристических объектов и т.д.   
5. Не стоит стремиться к открытию первого заведения с большим количеством посадочных мест. Рациональным подходом будет открытие одного небольшого заведения и, спустя определенное время, после некоторой стабилизации бизнеса и отработки бизнес-процессов, открытие следующих. При развитии бизнеса в сеть можно будет рассматривать открытие заведений разных типов и разным количеством посадочных мест.  

В качестве заключения можно сказать, что роботы в качестве официантов является отличным ходом для привлечения посетителей, а увеличить количество гостей позволит вкусная еда и приятная атмосфера.
