## Введение

**Цель:** выполнить предварительный анализ тарифов "Смарт" и "Ультра" компании "Мегалайн" - федерального оператора сотовой связи.

---

---   
**Основная задача:** проанализировать и определить, какой тариф приносит больше денег, чтобы скорректировать рекламный бюджет.

---    
**Данные:** выборка клиентов из 500 пользователейи за 2018 год, которая содержит сведения о том кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил.

Таблица **data_users** (информация о пользователях):  
`user_id` - уникальный идентификатор пользователя  
`first_name` - имя пользователя  
`last_name` - фамилия пользователя  
`age` - возраст пользователя (годы)  
`reg_date` - дата подключения тарифа (день, месяц, год)  
`churn_date` - дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)  
`city` - город проживания пользователя  
`tarif` - название тарифного плана  


Таблица **data_calls** (информация о звонках):   
`id` - уникальный номер звонка  
`call_date` - дата звонка  
`duration` - длительность звонка в минутах  
`user_id` - идентификатор пользователя, сделавшего звонок  


Таблица **data_messages** (информация о сообщениях):  
`id` - уникальный номер сообщения  
`message_date` - дата сообщения  
`user_id` - идентификатор пользователя, отправившего сообщение  


Таблица **data_internet** (информация об интернет-сессиях):  
`id` - уникальный номер сессии  
`mb_used` - объём потраченного за сессию интернет-трафика (в мегабайтах)  
`session_date` - дата интернет-сессии  
`user_id` - идентификатор пользователя  

Таблица **data_tariffs** (информация о тарифах):  
`tariff_name` - название тарифа  
`rub_monthly_fee` - ежемесячная абонентская плата в рублях  
`minutes_included` - количество минут разговора в месяц, включённых в абонентскую плату  
`messages_included` - количество сообщений в месяц, включённых в абонентскую плату  
`mb_per_month_included` - объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)  
`rub_per_minute` - стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)  
`rub_per_message` - стоимость отправки сообщения сверх тарифного пакета  
`rub_per_gb` - стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

```python
# Функция расчета аналитических статистик
def statistics(data, metric):
    data_smart = data[data['tariff'] == 'smart']
    mean_smart = np.mean(data_smart[metric])
    variance_smart = np.var(data_smart[metric], ddof=1)
    std_smart = np.std(data_smart[metric])
    median_smart = np.median(data_smart[metric])
    Q1_smart = np.percentile(data_smart[metric], 25)
    Q3_smart = np.percentile(data_smart[metric], 75)
    
    data_ultra = data[data['tariff'] == 'ultra']
    mean_ultra = np.mean(data_ultra[metric])
    variance_ultra = np.var(data_ultra[metric], ddof=1)
    std_ultra = np.std(data_ultra[metric])
    median_ultra = np.median(data_ultra[metric])
    Q1_ultra = np.percentile(data_ultra[metric], 25)
    Q3_ultra = np.percentile(data_ultra[metric], 75)
    
    print('Среднее значение для тарифа "Смарт":', round(mean_smart, 2),  
          '\nДисперсия для тарифа "Смарт":', round(variance_smart, 2),
          '\nСтандартное отклонение для тарифа "Смарт":', round(std_smart, 2),
          '\nМедиана для тарифа "Смарт":', round(median_smart, 2),
          '\nИнтерквартильный диапазон для тарифа "Смарт":', 
          round(Q1_smart, 2), '-', round(Q3_smart, 2),
          
          '\n\nСреднее значение для тарифа "Ультра":', round(mean_ultra, 2), 
          '\nДисперсия для тарифа "Ультра":', round(variance_ultra, 2), 
          '\nСтандартное отклонение для тарифа "Ультра":', round(std_ultra, 2),   
          '\nМедиана для тарифа "Ультра":', median_ultra,
          '\nИнтерквартильный диапазон для тарифа "Ультра":', 
          round(Q1_ultra, 2), '-', round(Q3_ultra, 2))
```

```python
# Функция для отрисовки графиков для анализа тарифов
def graph_hist(data, metric, bins, title, xlabel):
    fig_metric = plt.figure(figsize=(10, 4))
    ax_metric = sns.histplot(data, x=metric, hue='tariff', kde=True, bins=bins)
    ax_metric.legend(title='Тариф', bbox_to_anchor=(1, 1), labels=['Ультра', 'Смарт'])
    ax_metric.set_title(f'Гистограмма {title}', fontweight='bold', pad=14)
    ax_metric.set_xlabel(xlabel, fontweight='bold')
    ax_metric.set_ylabel('Частота', fontweight='bold')
    plt.show()
```

```python
# Функция для расчета разницы между использованными ресурсами и лимитом
def tariff_limit(data, metric, limit_smart, text1, limit_ultra, text2):
    print(f'Количество пользователей, превысивших лимит {text1} по тарифу "Смарт":', '{:.1%}'\
          .format(len(data[(data['tariff'] == 'smart') & (data[metric] > limit_smart)]) / 
                  len(data[data['tariff'] == 'smart'])),
          f'\nКоличество пользователей, использовавших меньше {text2} по тарифу "Ультра":', '{:.1%}'\
          .format(len(data[(data['tariff'] == 'ultra') & (data[metric] < limit_ultra)]) /
                  len(data[data['tariff'] == 'ultra'])))
```

### Группировка данных

Добавим в таблицу ***data_users*** столбец месяц и создадим новую таблицу сгруппировав по пользователям и месяцам количество звонков и их длительность:

```python
data_calls['month'] = data_calls['call_date'].dt.month
user_calls = data_calls.groupby(['user_id', 'month']).agg({'duration':'sum', 'dur_min':'sum',
'id':'count'}).reset_index().rename(columns={'id':'calls_count'})
user_calls.head(3)
```

Добавим в таблицу ***data_messages*** столбец месяц и создадим новую таблицу сгруппировав по пользователям и месяцам количество сообщений:

```python
data_messages['month'] = data_messages['message_date'].dt.month
user_messages = data_messages.groupby(['user_id', 'month']).agg({'id':'count'})\
.reset_index().rename(columns={'id':'messages_count'})
user_messages.head(3)
```

Добавим в таблицу ***data_internet*** столбец месяц и создадим новую таблицу сгруппировав по пользователям и месяцам количество сессий и общий трафик:

```python
data_internet['month'] = data_internet['session_date'].dt.month
user_internet = data_internet.groupby(['user_id', 'month']).agg({'mb_used':'sum', 'id':'count'})\
.reset_index().rename(columns={'id':'session_count'})
user_internet.head(3)
```

Преобразуем объем использованного трафика в соответствии с условиями тарификации:

```python
user_internet['gb_used'] = (user_internet['mb_used'] / 1024).apply(np.ceil).astype(int)
user_internet = user_internet.reindex(columns=['user_id', 'month', 'mb_used', 'gb_used', 
                                               'session_count'])
print('Тип данных переменной gb_used:', user_internet['gb_used'].dtypes)
```

### Расчет выручки

Напишем формулу для расчета выручки:

```python
def profit_def(row):
    if row['dur_min'] > row['minutes_included']:
        f1 = (row['dur_min'] - row['minutes_included']) * row['rub_per_minute']
    else:
        f1 = 0
    if row['messages_count'] > row['messages_included']:
        f2 = (row['messages_count'] - row['messages_included']) * row['rub_per_message']
    else:
        f2 = 0    
    if row['gb_used'] > row['gb_per_month_included']:
        f3 = (row['gb_used'] - row['gb_per_month_included']) * row['rub_per_gb']
    else:
        f3 = 0
    profit = row['rub_monthly_fee'] + f1 + f2 + f3
    return profit
data['profit'] = data.apply(profit_def, axis = 1)
data['profit'] = data['profit'].astype(int)
```

Таким образом, рациональным является применение непараметрических методов статистики. 

**Выдвенем гипотезы:**  
Нулевая гипотеза (H0) - средняя выручка пользователей тарифов «Смарт» и «Ультра» не различается.  
Альтернативная гипотеза (H1) - средняя выручка пользователей тарифов «Смарт» и «Ультра» различается.  
  
Проверку гипотез осуществим с помощью непараметрического статистического критерия оценки различий между двумя независимыми выборками Манна — Уитни (U-критерий). 

```python
# Mann-Whitney U test
smart = data[data['tariff'] == 'smart']
ultra = data[data['tariff'] == 'ultra']
stat, p = mannwhitneyu(smart['profit'], ultra['profit'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Средняя выручка пользователей тарифов «Смарт» и «Ультра» не различается (Нулевая гипотеза H0 не отвергнута)')
else:
    print('Средняя выручка пользователей тарифов «Смарт» и «Ультра» различается (Нулевая гипотеза H0 отвергнута)')
```

Выручка пользователей тарифа "Ультра" находится преимущественно в пределах абонентской платы (1950[1950 - 1950]) и выше в 1,9 раз, чем выручка пользователей тарифа "Смарт" (1023[550-1750]). При этом средняя выручка пользователей тарифа "Смарт" в 1,86 раз превышает абонентскую плату по данному тарифу.
