## Введение

**Цель** - исследовать надежность заемщиков банка.  

---
**Задачи:**  
* Определить есть ли зависимость между наличием детей и возвратом кредита в срок.
* Определить есть ли зависимость между семейным положением и возвратом кредита в срок.  
* Определить есть ли зависимость между уровнем дохода и возвратом кредита в срок.  
* Проанализировать как разные цели кредита влияют на его возврат в срок.


---
**Данные** - статистика о платежеспособности клиентов.  Предоставлены заказчиком - кредитным отделом банка, в виде файла ***data.csv:***  
* `children` - количество детей в семье
* `days_employed` - общий трудовой стаж в днях
* `dob_years` - возраст клиента в годах
* `education` - уровень образования клиента
* `education_id` - идентификатор уровня образования
* `family_status` - семейное положение
* `family_status_id` - идентификатор семейного положения
* `gender` - пол клиента
* `income_type` - тип занятости
* `debt` - имел ли задолженность по возврату кредитов
* `total_income` - ежемесячный доход
* `purpose` - цель получения кредита

---
**Ход исследования:**
1. Изучить предоставленные заказчиком данные.  
2. Осуществить предобработку данных:  
    * определить и заполнить пропущенные значения;  
    * определить и преобразовать тип данных;  
    * исправить ошибки в наборе данных;  
    * обработать дубликаты;  
    * провести лематизацию;

3. Провести анализ данных, с учетом задач исследования.  
4. Сформулировать выводы.

## Импорт, изучение и преобразование данных

### Импорт и изучение данных

Откроем файл с данными ***data.csv*** и изучим общую информацию:

```python
data = pd.read_csv('data.csv')
```

```python
display(data.head(3))
data.info()
```

Таблица ***data*** содержит 21525 строк и 12 столбцов. Наименования столбцов оставлены без изменений. Набор данных содержит различные артефакты и аномалии, требующие дальнейшей обработки. Столбцы `days_employed` (общий трудовой стаж в днях) и `total_income` (ежемесячный доход) содержат пропуски.

### Обработка пропусков

Столбцы `days_employed` (общий трудовой стаж в днях) и `total_income` (ежемесячный доход) содержат пропуски. Определим их количество, чтобы определиться с методом устранения:

```python
data_na = pd.DataFrame({'sum_na': data.isna().sum(), 'percent_na': data.isna().sum() / len(data)})
data_na['percent_na'] = data_na['percent_na'].map(lambda x: '{:.1%}'.format(x))
data_na.sort_values(by='percent_na', ascending=False)
```

Проверим, связаны ли между собой пропуски данных в столбцах общий трудовой стаж в днях и ежемесячный доход:

```python
count = 0
for i in range(len(data)):
    if pd.isna(data.loc[i, 'days_employed']) == pd.isna(data.loc[i, 'total_income']) == True:
        count += 1
print('Количество клиентов, которые не указали сведения о трудовом стаже и доходах:', count)
```

2174 клиента не предоставили данные о своем стаже работе и доходах, что составляет 10,1% выборки.
Не будем исключать эти данные из расчетов и заполним пропуски.

###### Обработка пропусков в столбце "общий трудовой стаж"

Трудовой стаж не может быть отрицательным числом, поэтому преобразуем его: 

```python
# Замена знака
data['days_employed'] = data['days_employed'].abs()
# Замена пропущенных значений на 0
data['days_employed'] = data['days_employed'].fillna(0)
```

Для удобства переведем значения трудового стажа из дней в годы (грубое преобразование):

```python
data['days_employed'] = round((data['days_employed'] / 365), 2)
```

Изучим некоторые статистические характеристики общего трудового стажа после преобразования:

```python
pd.DataFrame(data['days_employed'].describe().map(lambda x: '{:.2f}'.format(x)))
```

Кроме пропущенных значений, у некоторых заемщиков трудовой стаж превышает 1100 лет. Точно не известно как были получены данные и нет возможности выяснить происхождение значений. Сгруппируем данные по типу занятости и переведем из предполагаемых дней в годы, чтобы определить группы, в которой присутствуют неадекватные значения:

```python
data_days_employed = data[['income_type', 'days_employed']]
data_days_employed_grouped = pd.DataFrame()
data_days_employed_grouped['days_employed_median'] = data_days_employed.groupby('income_type')['days_employed'].median()
data_days_employed_grouped.sort_values('days_employed_median', ascending = False)
```

Возможно данные о трудовом стаже пенсионеров и безработных предоставлены в часах. Поскольку данный признак не был отобран в качестве исследуемого, дальнейшие пребразования нерациональны.  

###### Обработка пропусков в столбце "ежемесячный доход"

В столбце ежемесячный доход присутствуют пропущеные значения. Произведем группировку по типу деятельности клиентов (создадим словарь). Поскольку доход может значительно варьировать в разных группах, вычислим медиану для каждой группы:

```python
data_income_type = data[['income_type', 'total_income']]
data_income_type_dict = pd.DataFrame()
data_income_type_dict['total_income_median'] = data_income_type.groupby('income_type')['total_income']\
                                                               .median().map(lambda x: '{:.2f}'.format(x))
data_income_type_dict.sort_values('total_income_median', ascending = False)
```

Заменим пропуски медианой, соответствующей группе по типу занятости. Преобразуем строковые данные в численный тип:

```python
data['total_income'] = data['total_income'].fillna(data['income_type']\
                                                   .map(data_income_type_dict['total_income_median']))
data['total_income'] = pd.to_numeric(data['total_income'], downcast='unsigned')
print('Количество пропущенных значений в столбце ежемесячный доход:', 
      len(data[data['total_income'].isnull() == True]))
```

Пропущенные значения о ежемесячном доходе были восстановлены посредством заполнения медианой, вычисленной для разных групп занятости.

### Категоризация данных

#### Определение целей получения кредита

Иучим цели, на которые берут займы клиенты банка:

```python
data['purpose'].unique()
```

Для одинаковых целей были использованы разные формулировки, в тексте также присутсвуют опечатки.  

##### Лемматизация

С помощью лемматизации преобразуем текст и объединим похожие цели.  

Создадим функцию, которая возвращает список с ключевыми словами:

```python
m = Mystem()
def lemmas_purpose(purpose):
    lemmas = '' .join(m.lemmatize(purpose))
    return lemmas
```

Добавим столбец `purpose_words` в таблицу `data`, в который будут записываться отобранные слова и отбразим уникальные значения для столбца `purpose_words`:

```python
data['purpose_words'] = data['purpose'].apply(lemmas_purpose)
```

Посчитаем количество ключевых слов:

```python
Counter(data['purpose_words'])
```

С помощью лемматизации были определены ключевые слова (основы слов) и получен список с основными целями займов.

##### Формирование категорий с целями кредита

Сформируем несколько категорий объеденив наиболее частые цели получения кредита. Добавим в таблицу **data** столбец `purpose_group` с новыми категориями на основании результатов лемматизации:

```python
def purpose_group(purpose):
    if 'образ' in purpose:
        return 'образование'
    if 'свад' in purpose:
        return 'свадьба'
    if 'авто' in purpose:
        return 'автомобиль'
    if 'недвиж' or 'жил' in purpose:
        return 'недвижимость'

data['purpose_group'] = data['purpose_words'].apply(purpose_group)
```

Оценим результат категоризации:

```python
data_purpose = pd.DataFrame({'purpose_group' : data['purpose_group'].value_counts(), 'percent_purpose_group' : data['purpose_group'].value_counts() / len(data)})
data_purpose['percent_purpose_group'] = data_purpose['percent_purpose_group'].map(lambda x: '{:.2%}'.format(x))
data_purpose.sort_values(by='percent_purpose_group', ascending=False)
```

Чаще всего заемщики берут кредит для различных операциий с недвижимостью (жилой и коммерческой), в том числе на строительство и ремонт  - 50,39%, на покупку автомобиля - 20,07%, образование - 18,71% и свадьбу - 10,83%.

#### Формирование категорий "Образование" и "Семейное положение" 

В файле ***data*** присутствуют столбцы `education_id` и `family_status_id`, которые являются идентификаторами. Используем их для создания словарей "Образование" и "Семейное положение".

```python
education_dict = data[['education_id', 'education']]
education_dict = education_dict.drop_duplicates().reset_index(drop=True)
education_dict.sort_values('education_id')
```

```python
family_status_dict = data[['family_status_id', 'family_status']]
family_status_dict = family_status_dict.drop_duplicates().reset_index(drop=True)
family_status_dict.sort_values('family_status_id')
```

Мы сформировали словари "Образование" и "Семейное положение".

Для дальнейшего анализа были сформированы категории "Цели кредита", "Образование" и "Семейное положение". 

<a id="analiz_data"></a><h2>Анализ данных</h2>

### Определение зависимости между наличием детей и возвратом кредита в срок

Сформируем группы заемщиков, в зависимости от наличия и количества детей, и расчитаем долю тех, у кого есть задолжность.

```python
data_pv_child = pd.pivot_table(data, values='debt', index='children', aggfunc=['count','sum'])
data_pv_child['ratio'] = data_pv_child['sum'] / data_pv_child['count']
data_pv_child['ratio'] = data_pv_child['ratio'].map(lambda x: '{:.2%}'.format(x))
data_pv_child = data_pv_child.sort_values(by='ratio', ascending=False)
data_pv_child.columns = ['lenders', 'debt', 'ratio']
data_pv_child
```

Чаще задолжность возникает у семей с 4 (9,76%), 2 (9,49%) и 1 (9.17%) детьми. Несколько реже у клиентов с 3 (8,18%) и еще реже у бездетных семей (7,54%). В категории семьи с 5 детьми задолжность не возникала. Данное значение вероятно связано с малым количеством наблюдений.

### Определение зависимости между семейным положением и возвратом кредита в срок

Расчитаем доли должников, в зависимости от семейного положения.

```python
data_pv_fam = pd.pivot_table(data, values='debt', index='family_status', aggfunc=['count','sum'])
data_pv_fam['ratio'] = data_pv_fam['sum'] / data_pv_fam['count']
data_pv_fam['ratio'] = data_pv_fam['ratio'].map(lambda x: '{:.2%}'.format(x))
data_pv_fam = data_pv_fam.sort_values(by='ratio', ascending=False)
data_pv_fam.columns = ['lenders', 'debt', 'ratio']
data_pv_fam
```

Не женатые (9,75%) и состоящие в гражданском браке (9,35%) чаще имеют задолжность по кредиту, против тех кто состоит в браке (7,55%) или развелся (7,11%). Вдовствующие не выплачивают займы в срок только в 6,57% случаев.  

Посмотрим на медианный возраст в группах с разным сеймейным статусом и дополнительной группировкой по наличию задолжности.

```python
data_years = pd.pivot_table(data, index=['family_status', 'debt'], values = ['dob_years', 'total_income'],
                            aggfunc= {'dob_years':['median','count'], 'total_income':'median'})
data_years.columns = ['lenders', 'age_median', 'total_income_median']
data_years.astype(int)
```

Среди вдовствующих закономерно оказались клиенты старше медианного возраста. Обращает на себя внимание, что доход в группах примерно одинаковый и не зависит от семейного положения. 

Изучим цели, на которые берут займ разные когорты клиентов:

```python
new_table = data[['family_status', 'purpose_group']]
new_table = new_table.reindex(columns=new_table.columns.tolist() + ['col'])
new_table['col'] = new_table['col'].fillna(1).astype(int)
data_coef = pd.pivot_table(new_table, index='family_status', columns='purpose_group', values = 'col',
                           fill_value = 0, aggfunc=sum).apply(lambda x: (x / x.sum())*100, 
                                                              axis=1).round(2).astype(str) + '%'
data_coef
```

Мы провели исследование и выяснили, что чаще всего клиенты, живущие в гражданском браке берут кредит на свадьбу. В остальных группах в одинаковой степени заинтересованы недвижимостью.
Таким образом, не получилось найти значимый фактор выделящий вдовствующих клиентов среди других. Возможно высокий процент возврата кредитов в данной группе связан с возрастом, который выше медианного значения и может коррелировать с большим жизненным опытом и ответственной позицей в отношении долгов.  

### Определение зависимости между уровнем дохода и возвратом кредита в срок

Изучим некоторые статистические характеристики ежемесячного дохода.

```python
pd.DataFrame(data['total_income'].describe().map(lambda x: '{:.1f}'.format(x)))
```

Сформируем три группы клиентов по уровню доходов (низкий, средний и высокий).

```python
data['total_income_group'] = pd.qcut(data['total_income'], 3, labels=['низкий', 'средний', 'высокий'])
```

Расчитаем долю должников среди клиентов банка, в зависимости от уровня доходов.

```python
data_pv_inc = pd.pivot_table(data, values='debt', index='total_income_group', aggfunc=['count','sum'])
data_pv_inc['ratio'] = data_pv_inc['sum'] / data_pv_inc['count']
data_pv_inc['ratio'] = data_pv_inc['ratio'].map(lambda x: '{:.2%}'.format(x))
data_pv_inc = data_pv_inc.sort_values(by='ratio', ascending=False)
data_pv_inc.columns = ['lenders', 'debt', 'ratio']
data_pv_inc
```

Заемщики с доходами выше медианных значений (7,52%) реже становятся должниками, чем с доходами на уровне медианы (8,68%) и ниже медианных значений (8,12%). 

### Определение зависимости между целями кредита и возвратом кредита в срок

Оценим долю клиентов банка с задолжностью по кредиту, в зависимости от целей займа.

```python
data_pv_pur = pd.pivot_table(data, values='debt', index='purpose_group', aggfunc=['count','sum'])
data_pv_pur['ratio'] = data_pv_pur['sum'] / data_pv_pur['count']
data_pv_pur['ratio'] = data_pv_pur['ratio'].map(lambda x: '{:.2%}'.format(x))
data_pv_pur = data_pv_pur.sort_values(by='ratio', ascending=False)
data_pv_pur.columns = ['lenders', 'debt', 'ratio']
data_pv_pur
```

Чаще формируются задолжность у тех кто брал займ на автомобиль (9,36%) и образование (9,22%), реже - на свадьбу (8,0%) и операции с недвижимостью (7,23%).
