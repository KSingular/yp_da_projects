## Введение

<b><font color="#4682B4">Цель</font></b> - провести мониторинг клиентов провайдера виртуальной телефонии "Нупозвони", которые экономически неэффективно используют услуги на своих тарифах и могут уменьшить свои расходы, перейдя на тариф с более выгодными условиями. Перевод клиентов на выгодные тарифы поможет снизить вероятность их оттока. 

Клиентами "Нупозвони" являются организации (юр. лица), которым нужно распределять множество входящих вызовов на различных операторов, либо осуществлять исходящие вызовы силами операторов. Также операторы могут осуществлять внутренние вызовы, то есть вызовы друг между другом, которые осуществляются внутри сети виртуальной телефонии.

---
<b><font color="#4682B4">Заказчик исследования:</font></b> коммерческий департамент.

---
<b><font color="#4682B4">Задачи проекта:</font></b>  

**Установить сколько клиентов используют экономически невыгодные тарифы и как изменится выручка компании, при переводе всех клиентов на оптимальные тарифы:**
 - определить клиентов, которые неэффективно используют услуги провайдера виртуальной связи;
 - изучить динамику работы клиентов телеком провайдера и выделить активных клиентов, по которым будут осуществляться расчеты;
 - изучить текущую выручку компании (в разрезе месяца), а также динамику за весь период среднего чека в разрезе тарифа и определить уровень переплаты;
 - изучить потери в выручке, если все клиенты перейдут на выгодные тарифы;
 - предложить пути решения проблемы потери выручки.

Исследование позволит лучше понять как пользователи потребляют услуги и, возможно, выдвинуть гипотезы нового ценообразования.  

***Планируется проводить такие исследования раз в квартал.***

<b><font color="#4682B4">План исследования:</font></b>  

1. Загрузить данные и подготовить их к исследованию: переимовать столбцы, преобразовать типы данных, устранить пропуски, дубликаты, аномальные значения и т.д.. Объяснить их появление.  
2. Провести исследовательский анализ данных:

     * определить основные профили клиентов по использованию услуг;
     * определить затраты пользователей на текущем тарифе;
     * определить клиентов, которые неэффективно используют услуги провайдера виртуальной связи; 
     * определить гипотетические затраты пользователей при переходе на другой тариф.
 * Изучить текущую выручку компании (в разрезе месяца), а также динамику за весь период среднего чека в разрезе тарифа и определить уровень переплаты. Для чего произвести расчет метрик и бизнес-показателей (WAU, Retention Rate, Churn Rate, средний чек, LTV).
 * Изучить потери в выручке, если все клиенты перейдут на выгодные тарифы.  
 * Найти решения по снижению потери в выручке.
3. Проверить статистические гипотезы:  
    * Гипотеза № 1. Выручка провайдера не изменится при переводе клиентов на другие тарифы.
    * Гипотеза № 2. Уровень дополнительного дохода за счет использования клиентами неоптимального тарифа не отличается для разных тарифных планов.
    * Гипотеза № 3. Уровень оттока клиентов при использовании оптимального и неоптимального тарифного плана не отличается.
4. Сформулировть выводы по каждому разделу исследования и общий вывод. 
5. Подготовить презентацию для директора коммерческого департамента с результатами исследования и рекомендациями.
6. Подготовить отчет для менеджеров коммерческого департамента, по данным которого будет проведена работа с клиентами (предоставить список клиентов, которые неэффективно для себя пользуются услугами Телекома и которым целесообразно предложить смену тарифного плана). 
7. Построить дашборды иллюстрирующие потребление услуг клиентами Телекома за весь период времени в разрезе тарифного плана (с возможностью фильтрации по типу звонка). 
    1. Построить диаграмму количества вызовов по дням.
    2. Построить круговую диаграмму, отображающую соотношение кол-ва входящих и исходящих вызовов.
    3. Добавить фильтр дашборда по типу вызова (внешний или внутренний).

## Правила формирования тарифов

**Расходы клиента состоят из:**  
* ежемесячной платы +
* платы за перерасход по внутренним исходящим звонкам, **с учетом времени ожидания** +
* платы за внешние исходящие звонки, **с учетом времени ожидания** +
* платы 100 руб за оператора  

| Тарифный план | Ежемесячная плата | Входящие | <div style="width:250px">Исходящие звонки внутри сети - бесплатны до 2000 минут в месяц на всех тарифах. Цена сверх лимита:</div> | <div style="width:150px">Исходящие звонки, кроме внутренних:</div> |
|:---:|:---:|:---:|:---:|:---:|
|Тариф А| 5000 ₽ | 0 | 0.1 ₽/мин. | 0.4 ₽/мин. |
|Тариф B| 2000 ₽ | 0 | 0.15 ₽/мин. | 0.5 ₽/мин. |
|Тариф C| 1000 ₽ | 0 | 0.3 ₽/мин. | 0.7 ₽/мин. |  

Счета выставляются на прошедший календарный месяц, поэтому списание можно считать от 1-го числа каждого месяца. Остаток трафика сгорает на начало месяца. Например, клиент начал платить 9-го числа, это значит что за период с 9 по конец месяца у него работают действующие аккумуляторы трафика (например, 2000 бесплатных внутренних минут). Тарификация происходит на основе кумулятивных показателей за период расчета (календарный месяц). Продолжительность звонков за месяц округляется в большую сторону. Например, у клиента за месяц было 2 звонка, если 1 января был разговор на 15
секунд, а 2 января разговор на 7 секунд, то в сумме за январь было 22 секунды. Округляем до 1 минуты.

## Описание данных

Датасеты содержат данные об использовании услуг виртуальной телефонии "Нупозвони". 

Колонки в **telecom_clients.csv:**

- `user_id` - идентификатор клиента
- `tariff_plan` - тарифный план клиента
- `date_start` - дата регистрации клиента в сервисе    

Колонки в **telecom_dataset.csv**:

- `user_id` - идентификатор клиента
- `date` - дата статистики
- `direction` - направление вызовов: 
    - `out` - исходящий вызов 
    - `in` - входящий
- `internal` - маркер внутренних и внешних вызовов  
    Вызовы между сотрудниками одного колл-центра называют внутренними, остальные - внешними
- `operator_id` - идентификатор оператора
- `is_missed_call` - маркер пропущенных вызовов
- `calls_count` - количество звонков
- `call_duration` - длительность звонка без учёта времени ожидания
- `total_call_duration` - длительность звонка с учётом времени ожидания

```python
# Функция аннотирования медианы
def add_median_labels(ax, precision='.0f'):
    lines = ax.get_lines()
    boxes = [c for c in ax.get_children() if type(c).__name__ == 'PathPatch']
    lines_per_box = int(len(lines) / len(boxes))
    for median in lines[4:len(lines):lines_per_box]:
        x, y = (data.mean() for data in median.get_data())
        # choose value depending on horizontal or vertical plot orientation
        value = x if (median.get_xdata()[1] - median.get_xdata()[0]) == 0 else y
        text = ax.text(x, y, f'{value:{precision}}', ha='center', va='center',
                       color='white', fontweight='bold', size=8, bbox=dict(facecolor='#445A64'))
```

```python
# Добавление стилей к графикам
sns.set_palette('Dark2')
sns.set_style('darkgrid', {'patch.edgecolor': 'black', 'axes.edgecolor': 'black'})
```

```python
# Диаграммы размаха метрик direction_in, internal
def boxplot_def():
    fig = plt.figure(figsize=(16,3.5))
    for i, metric in enumerate(['calls_count', 'calls_duration', 'total_calls_duration']):
        ax_metrics_calls = fig.add_subplot(1,3,i+1)
        ax_metrics_calls = sns.boxplot(y=df_dataset[metric])
        ax_metrics_calls.margins(0.05, 0.05)
        plt.title('Диаграммы размаха метрики\n' + metric, fontweight='bold', pad=14) 
        plt.ylabel('')
        add_median_labels(ax_metrics_calls)
    plt.show()
```

## Группировка данных и расчет дополнительных метрик

Учитывая, что счета выставляются за прошедший календарный месяц, выделим в отдельные столбцы сведения о месяце начала пользования тарифом и месяце получения метрики:

#### Расчет Retention Rate (коэффициент удержания)

Сформируем когорты и рассчитаем **Retention Rate** (коэффициент удержания):

```python
sns.set_style('white')
```

```python
# Формирование когорт по месяцу первой активности, в зависимости от времени жизни когорт
df_services_pt['cohort_lifetime'] = df_services_pt['month_metric'] - df_services_pt['month_start']
df_services_pt['cohort_lifetime'] = df_services_pt['cohort_lifetime'] / np.timedelta64(1, 'M')
df_services_pt['cohort_lifetime'] = df_services_pt['cohort_lifetime'].round().astype('int') + 1
df_services_pt['month_start'] = df_services_pt['month_start'].dt.strftime('%Y-%m')
cohorts_ret = df_services_pt.pivot_table(index='month_start', columns='cohort_lifetime', 
                                         values='user_id', aggfunc='nunique')
retention = pd.DataFrame()
for col in cohorts_ret.columns:
    retention = pd.concat([retention, cohorts_ret[col] / cohorts_ret[1]], axis=1)
retention.columns = cohorts_ret.columns
retention.loc['Среднее'] = retention.mean()
fig_ret = plt.figure(figsize=(5, 3))
ax_ret = sns.heatmap(retention.loc[:, 2:], annot=True, fmt='.1%', linewidths=1, linecolor='black', 
                     cmap='coolwarm', center=0, vmin=0, vmax=1)
cbar = ax_ret.collections[0].colorbar
cbar.set_ticks([0, .25, .5, .75, 1])
cbar.set_ticklabels(['0', '20%', '50%', '75%', '100%'])
ax_ret.set_title('Retention Rate', fontweight='bold', pad=14)
ax_ret.set_xlabel('Время жизни когорт')
ax_ret.set_ylabel('Когорты (месяц подключения)')
plt.yticks(rotation=0) 
plt.show()
```

**Retention Rate** (коэффициент удержания) клиентов достаточно высокий, но начал снижается, что говорит об оттоке клиентов, несмотря их значимый приток.

#### Расчет Churn Rate  (коэффициент оттока)

Рассчитаем общий процент оттока за 3 месяца:

```python
user_all = df_services_pt['user_id'].nunique()
user_max = df_services_pt[df_services_pt['month_metric'] == df_services_pt['month_metric']\
                          .max()]['user_id'].nunique()
print('Количество клиентов, которые ушли в отток за последние 3 месяца:', (user_all - user_max), 
      '- {:.1%}'.format((user_all - user_max) / user_all))
```

Сформируем когорты и рассчитаем **Churn Rate** (коэффициент оттока):

```python
# Формирование когорт по месяцу первой активности, в зависимости от времени жизни когорт
cohorts_churn = df_services_pt.groupby(['month_start', 'cohort_lifetime'])\
                              .agg({'user_id': 'nunique'}).reset_index()
initial_users_count = cohorts_churn[cohorts_churn['cohort_lifetime'] == 1][['month_start', 'user_id']]
initial_users_count = initial_users_count.rename(columns={'user_id': 'cohort_users'})
cohorts_churn = cohorts_churn.merge(initial_users_count, on='month_start')
cohorts_churn['churn_rate'] = cohorts_churn.groupby(['month_start'])['user_id'].pct_change()
churn_pivot = cohorts_churn.pivot_table(index='month_start', columns='cohort_lifetime', 
                                        values='churn_rate', aggfunc='sum')
churn_pivot.loc['Среднее'] = churn_pivot.mean()
fig_ret = plt.figure(figsize=(5, 3))
ax_churn = sns.heatmap(churn_pivot.loc[:, 2:], annot=True, fmt='.1%', linewidths=1, linecolor='black', 
                     cmap='coolwarm', center=0, vmin=-0.2, vmax=0.2) 
cbar = ax_churn.collections[0].colorbar
cbar.set_ticks([-0.2, -.1, 0, .1, 0.2])
cbar.set_ticklabels(['-20', '-10%', '0', '10%', '20%'])
ax_churn.set_title('Churn Rate', fontweight='bold', pad=14)
ax_churn.set_xlabel('Время жизни когорт')
ax_churn.set_ylabel('Когорты (месяц подключения)')
plt.yticks(rotation=0) 
plt.show()
```

**Churn Rate (коэффициент оттока)** говорит о том что значимый отток происходит на второй месяц использования услуг провайдера. Однако большинство клиентов сохраняют свою лояльность. Таким образом, второй месяц является критическим для возможности удержания клиента.

### Определение профилей клиентов (кластеризация)

Для определения и описания ключевых профилей клиентов выполним кластеризацию. Сформируем таблицу с аггрегированными данными числа операторов, а также средними значениями длительности и количеством входящих и исходящих звонков для каждого клиента:

```python
df_cluster = df_services_pt.groupby('user_id').agg({'n_operator' : lambda x: x.mode()[0]}).reset_index()\
.merge(df_services_pt.query('direction_in == True').groupby('user_id')\
.agg(in_calls_count = ('calls_count', 'mean'),
     in_calls_duration = ('calls_duration', 'mean'),
     in_t_calls_duration = ('total_calls_duration', 'mean')).reset_index(), on = 'user_id', how = 'left')\
.merge(df_services_pt.query('direction_in == False').groupby('user_id')\
.agg(out_calls_count = ('calls_count', 'mean'),
     out_calls_duration = ('calls_duration', 'mean'),
     out_t_calls_duration = ('total_calls_duration', 'mean')).reset_index(), on = 'user_id', how = 'left')\
.fillna(0)
```

#### Кластеризация

Определим переменные, которые будут участвовать в кластеризации:

```python
# Отбор переменных для кластеризаци - направление и маркер звонков,
# среднее количество операторов и средняя общая длительность звонков 
profiles = ['n_operator', 'in_t_calls_duration', 'out_t_calls_duration']
```

Стандартизируем данные:

```python
# Стандартизация данных
sc = StandardScaler() 
# Обучение нормализатора и преобразование набора данных
X_sc = sc.fit_transform(df_cluster[profiles])
linked = linkage(X_sc, method = 'ward')
```

Построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму:

```python
# Добавление стилей к графикам
sns.set_palette('Set1')
sns.set_style('white')
```

```python
# Построение матрицы расстояний
linked = linkage(X_sc, method = 'ward')
# Построение дендрограмы
plt.figure(figsize=(8, 3.5))  
dendrogram(linked, truncate_mode='lastp')
plt.title('Дендрограмма иерархической кластеризации клиентов', fontweight='bold', pad=14)
plt.xlabel('Размер кластера')
plt.ylabel('Расстояние')
plt.show() 
```

В соответствии с распределением цветов на дендрограмме определено 3 кластера. 

Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов:

```python
# Число кластеров, равное 3
km = KMeans(random_state=0, n_clusters = 3)  
# Применение алгоритма к данным и формирование вектора кластеров
labels = km.fit_predict(X_sc) 
# Сохранение меток кластера в поле датасета
df_cluster['cluster_km'] = labels
```

Посчитаем метрику силуэта для кластеризации, чтобы оценить качество модели:

```python
print('Метрика силуэта: {:.3f}'.format(silhouette_score(X_sc, labels)))
```

Качество модели выше среднего.

### Расчет гипотетической стоимости услуг связи в соответствии с тарификацией

Для более полного анализа расчеты произведем в том числе и для пользователей ушедших в отток.  

Создадим таблицу **df_tariffs** содержащую сведения о тарифах:

```python
df_tariffs = pd.DataFrame(
    {
        'tariff_plan': ['A', 'B', 'C'],
        'price_month': [5000, 2000, 1000],
        'calls_out_internal': [0.1, 0.15, 0.3],
        'calls_out_external': [0.4, 0.5, 0.7]
    }
)
```

```python
df_tariffs
```

В соответствии с тарифами продолжительность всех внутренних и внешних исходящих звонков с ожиданием суммируется за месяц и округляется в большую сторону. Преобразуем секунды в минуты и округлим полученные значения в большую сторону:

```python
df_services_pt['duration_min'] = np.ceil(df_services_pt['total_calls_duration'] / 60).astype('int64')
```

```python
df_services_pt.head(3)
```

Рассчитаем расходы на исходящие звонки внутренние и внешние для всех предполагаемых тарифов:

```python
for i, tariff in enumerate(df_tariffs['tariff_plan']):
    df_services_pt['part_' + tariff] = df_services_pt.loc[(df_services_pt['direction_in'] == False) &
                                                          (df_services_pt['internal'] == True) & 
                                                          (df_services_pt['duration_min'] > 2000)]['duration_min'] \
                                       * df_tariffs['calls_out_internal'][i]
    df_services_pt['part_' + tariff] = df_services_pt.loc[(df_services_pt['direction_in'] == False) &
                                                          (df_services_pt['internal'] == False)]['duration_min'] \
                                       * df_tariffs['calls_out_external'][i]
    df_services_pt['part_' + tariff] = df_services_pt['part_' + tariff].fillna(0)
```

Выполним сведение данных в таблицу **df_profit** и добавим к рассчитанным для тарифов расходам на исходящие внутренние и внешние звонки ежемесячную плату и расходы на операторов:

```python
# Сведение данных с вычислением суммы
df_profit = pd.pivot_table(df_services_pt, values=['part_A', 'part_B', 'part_C'], index=['user_id', 
                           'tariff_plan', 'month_start', 'month_metric', 'last_month', 'lifetime', 
                           'churn', 'cohort_lifetime', 'n_operator'], aggfunc='sum').reset_index()
```

```python
df_profit['user_id'] = df_profit['user_id'].astype('object')
```

```python
# Добавление сведений о ежемесячной плате и расходов на операторов
for i, tariff in enumerate(df_tariffs['tariff_plan']):
    df_profit['profit_' + tariff] = df_tariffs['price_month'][i] + (df_profit['n_operator'] * 100) \
    + df_profit['part_' + tariff]
df_profit = df_profit[['user_id', 'tariff_plan', 'month_start', 'month_metric', 'last_month', 'lifetime',
                       'cohort_lifetime', 'n_operator', 'churn', 'profit_A', 'profit_B', 'profit_C']]
```

```python
df_profit.head(3)
```

Таблица **df_profit** содержит данные о гипотетических затратах клиентов при использовании разных тарифов.

### Помесячный расчет актуальных затрат клиентов на услуги связи

Рассчитаем затраты клиентов на услуги связи для актульного тарифа (в том числе для пользователей ушедших в отток):

```python
def def_actual(row):
    if row['tariff_plan'] == 'A':
        return row['profit_A']
    if row['tariff_plan'] == 'B':
        return row['profit_B']
    if row['tariff_plan'] == 'C':    
        return row['profit_C']

df_profit['profit_actual'] = df_profit.apply(def_actual, axis = 1)
```

```python
df_profit.head(3)
```

Таблица **df_profit** содержит данные о гипотетических затратах клиентов при использовании разных тарифов и актуальных затратах клиентов.

### Определение выгодного тарифа каждого клиента и расчет переплаты

Выгода клиента зависит от затрат на звонки сверх лимита 2000 минут внутри сети и затрат на исходящие звонки и их оптимальном балансе с ежемесячной платой. Затраты на оператора являются индивидуальными. Определим среди тарифов наиболее выгодный для каждого клиента в каждом месяце использования услуг связи провайдера:

```python
# Определение выгодного тарифа
df_profit['best_tariff'] = df_profit[['profit_A', 'profit_B', 'profit_C']].idxmin(axis=1)
```

```python
# Переименование для возможности последующего сравнения тарифов
df_profit['best_tariff'] = df_profit['best_tariff'].replace('profit_A', 'A')
df_profit['best_tariff'] = df_profit['best_tariff'].replace('profit_B', 'B')
df_profit['best_tariff'] = df_profit['best_tariff'].replace('profit_C', 'C')
```

Рассчитаем гипотетические затраты при оптимальном тарифе:

```python
# Расчет гипотетических затрат
def def_hypothetic(row):
    if row['best_tariff'] == 'A':
        return row['profit_A']
    if row['best_tariff'] == 'B':
        return row['profit_B']
    if row['best_tariff'] == 'C':    
        return row['profit_C']

df_profit['profit_hypothetic'] = df_profit.apply(def_hypothetic, axis = 1)
```

Рассчитаем переплату для каждого клиента:

```python
df_profit['overpayment'] = df_profit['profit_actual'] - df_profit['profit_hypothetic']
```

Рассчитаем процент переплаты относительно гипотетического тарифа:

```python
df_profit['%overpayment'] = (df_profit['overpayment'] / df_profit['profit_hypothetic'] * 100).round(1)
```

```python
df_profit.head(3)
```

Таблица **df_profit** содержит данные о гипотетических затратах клиентов при использовании разных тарифов, актуальных затратах клиентов, гипотетических затратах при использовании оптимального тарифа и разницы между затратами при оптимальном и неоптимальном тарифах.

### Расчет размера среднего чека

Рассчитаем **средний чек**, применив группировку по месяцу подключения и использования тарифа:

```python
# Сводная таблица по тарифу и месяцу его использования (когорты)
revenue_per_client = df_active.groupby(['tariff_plan', 'month_metric'])\
                              .agg({'profit_actual':'sum', 'user_id':'nunique'}).reset_index()
revenue_per_client.columns = ['tariff_plan', 'month_metric', 'profit_actual', 'n_clients']
revenue_per_client['month_metric'] = revenue_per_client['month_metric'].dt.strftime('%Y-%m')
# Выручка на одного клиента
revenue_per_client['rev_per_client'] = (revenue_per_client['profit_actual'] / 
                                        revenue_per_client['n_clients']).round(2)
# Сводная таблица с расчетом среднего чека
revenue_per_client_month = revenue_per_client.pivot_table(index=['tariff_plan'], 
                                              columns='month_metric', values='rev_per_client', aggfunc='mean')
revenue_per_client_month['Среднее'] = revenue_per_client_month.mean(axis=1)
# Тепловая карта
fig_rev_m = plt.figure(figsize=(8,2))
ax_rev_m  = sns.heatmap(revenue_per_client_month, annot=True, fmt='.2f', linewidths=1, 
                        linecolor='black', cmap='coolwarm', vmin=0, vmax=8000)
ax_rev_m.set_title('Средний чек', fontweight='bold', pad=14)
ax_rev_m.set_xlabel('Месяц использования тарифа')
ax_rev_m.set_ylabel('Тариф')
plt.yticks(rotation=0) 
plt.show()
```

Самые большие **средние чеки** за 4 месяца отмечены для клиентов использующих тариф "А" - 6037,45 рублей. **Средний чек** клиентов использующих тариф "В" составляет 2510,15 рублей и для клиентов использующих тариф "С" - 1358,87 рублей.    

Отразим динамику **среднего чека** на линейном графике:

```python
# Добавление стилей к графикам
sns.set_palette('Set1')
```

```python
avg_check = df_active.groupby(['tariff_plan', 'month_metric'])\
                     .agg({'profit_actual':'sum', 'user_id':'nunique'}).reset_index()
avg_check.columns = ['tariff_plan', 'month_metric', 'profit_actual', 'n_clients']
avg_check['rev_per_client'] = (avg_check['profit_actual'] / avg_check['n_clients']).round(2)
fig=plt.figure(figsize=(12,4))
ax_avg_check = sns.lineplot(data=avg_check, x='month_metric', y='rev_per_client', hue='tariff_plan', marker='o')
for i, txt in enumerate(avg_check['rev_per_client']):
    ax_avg_check.annotate(txt, xy=(avg_check['month_metric'][i], txt+150), ha='center', va='bottom') 
plt.legend(title='Тариф', loc='upper left', labels=['A', 'B', 'C'])    
ax_avg_check.margins(0.05, 0.1)  
plt.title('Динамика среднего чека', fontweight='bold', pad=14) 
plt.xlabel('Месяц')
plt.ylabel('Средний чек')           
fig.show()
```

### Расчет LTV ("пожизненной" ценности клиента)

Произведем расчет **LTV** ("пожизненной" ценности клиента) для клиентов использующих разные тарифы при маржинальности  100%:

```python
# Расчет LTV
ltv_clients = df_active.groupby(['tariff_plan', 'cohort_lifetime'])\
                       .agg({'profit_actual':'sum', 'user_id':'nunique'}).reset_index()
ltv_clients.columns = ['tariff_plan', 'cohort_lifetime', 'profit_actual', 'n_clients']
margin_rate = 1
ltv_clients['gp'] = ltv_clients['profit_actual'] * margin_rate
ltv_clients['ltv'] = ltv_clients['gp'] / ltv_clients['n_clients']
output_ltv = ltv_clients.pivot_table(index=['tariff_plan'], columns='cohort_lifetime', 
                                       values='ltv', aggfunc='sum').round(2).cumsum(axis=1) 
output_ltv.loc['Среднее'] = output_ltv.mean()
# Тепловая карта
fig_output_ltv = plt.figure(figsize=(8, 2.8))
ax_output_ltv = sns.heatmap(output_ltv, annot=True, fmt='.2f', linewidths=1, linecolor='black', cmap='coolwarm',
                            vmin=0, vmax=30000) 
ax_output_ltv.set_title('LTV', fontweight='bold', pad=14)
ax_output_ltv.set_xlabel('Месяц использования тарифа')
ax_output_ltv.set_ylabel('Тариф')
plt.yticks(rotation=0)
plt.show()
```

За 4 месяца каждый клиент использующий тариф "А" приносит в среднем 24418,06 рублей, тариф "В" - 10268,19 рулей, тариф "С" - 5664,63 рублей.

## Проверка гипотез

### Гипотеза № 1. Выручка провайдера не изменится при переводе клиентов на другие тарифы

Расчет выполним для последнего месяца получения выручки, предположив, что клиенты воспроизведут похожие метрики в следующем месяце:

```python
profit_actual = df_profit[df_profit['month_metric'] == df_profit['month_metric'].max()]['profit_actual']
profit_hypothetic = df_profit[df_profit['month_metric'] == df_profit['month_metric'].max()]['profit_hypothetic']
```

```python
print('Выручка провайдера за последний месяц:', profit_actual.sum().round(1), 'рублей',
      '\nГипотетическая выручка провайдера за последний месяц', profit_hypothetic.sum(), 'рублей')
```

Выдвинем гипотезы:
* **Нулевая гипотеза (H0)** - выручка провайдера статистически значимо не изменится при переводе клиентов на другие тарифы
* **Альтернативная гипотеза (H1)** - выручка провайдера статистически значимо изменится при переводе клиентов на другие тарифы  

Изучим распределение переменных, чтобы определиться с критерием сравнения выборок:

```python
# Добавление стилей к графикам
sns.set_palette('Dark2')
```

```python
# Функция построения гистограмм 
def graph_hist(data, xvalue, title, xlabel):
    fig_revenue = plt.figure(figsize=(8,3.5))
    ax_revenue = sns.histplot(data, x=xvalue, kde=True)
    ax_revenue.set_title(title, fontweight='bold', pad=14)
    ax_revenue.set_xlabel(xlabel)
    ax_revenue.set_ylabel('Частота')
    plt.show()
```

```python
graph_hist(df_profit[df_profit['month_metric'] == df_profit['month_metric'].max()], 'profit_actual', 
           'Распределение выручки за последний месяц', 'Выручка за последний месяц')
```

```python
graph_hist(df_profit[df_profit['month_metric'] == df_profit['month_metric'].max()], 'profit_hypothetic', 
           'Распределение гипотетической выручки за последний месяц', 'Гипотетическая выручка за последний месяц')
```

Распределение переменных отличается от нормального. Проверку гипотез осуществим с помощью непараметрического статистического критерия оценки различий между двумя независимыми выборками Манна-Уитни (U-критерий).

```python
# Mann-Whitney U test
alpha = 0.05  # критический уровень статистической значимости
results = stats.mannwhitneyu(profit_actual, profit_hypothetic)
print('p-value: {0:.4f}'.format(results.pvalue))
if results.pvalue > alpha:
    print('Нулевая гипотеза H0 не отвергнута: выручка провайдера статистически значимо не изменится',
          'при переводе клиентов на другие тарифы')
else:
    print('Нулевая гипотеза H0 отвергнута: выручка провайдера статистически значимо изменится',
          'при переводе клиентов на другие тарифы')
```

* **Альтернативная гипотеза (H1)** - Уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически отличается для разных тарифных планов

Распределение переменных отличается от нормального. Проверку гипотез осуществим с помощью критерий Краскела-Уоллиса (Н-критерий) для проверки равенства медиан нескольких выборок. Апостериорные сравнения выполним с помощью непараметрического статистического критерия оценки различий между двумя независимыми выборками Манна-Уитни (U-критерий). Поправку на множественные сравнения выполним методом Бонферрони.

```python
# Расчет критерия Краскела-Уоллиса
stat, p = stats.kruskal(over_1, over_2, over_3)
print('Statistics=%.4f, p=%.4f' % (stat, p))
alpha = 0.05
bonferroni_alpha = alpha / 3  # произведено три сравнения
if p > alpha:
    print('Нулевая гипотеза H0 не отвергнута: уровень дополнительного дохода за счет использования', 
          'клиентами неоптимального тарифа статистически значимо не отличается для разных тарифных планов')
else:
    print('Нулевая гипотеза H0 отвергнута: уровень дополнительного дохода за счет использования', 
          'клиентами неоптимального тарифа статистически значимо отличается для разных тарифных планов')
```

Апостериорное сравнение уровня дополнительного дохода для тарифов "А" и "В":
* **Нулевая гипотеза (H0)** - уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически значимо не отличается между тарифами "А" и "В"
* **Альтернативная гипотеза (H1)** - уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически значимо отличается между тарифами "А" и "В"

```python
# Mann-Whitney U test
alpha = 0.05  # критический уровень статистической значимости
results = stats.mannwhitneyu(over_1, over_2)
print('p-value: {0:.4f}'.format(results.pvalue))
if results.pvalue < bonferroni_alpha:
    print('Нулевая гипотеза H0 отвергнута: уровень дополнительного дохода за счет использования', 
          'клиентами неоптимального тарифа статистически значимо отличается между тарифами "А" и "В"')
else:
    print('Нулевая гипотеза H0 не отвергнута: уровень дополнительного дохода за счет использования', 
          'клиентами неоптимального тарифа статистически значимо не отличается между тарифами "А" и "В"')
```

Апостериорное сравнение уровня дополнительного дохода для тарифов "В" и "С":

Апостериорное сравнение уровня дополнительного дохода для тарифов "А" и "C":
* **Нулевая гипотеза (H0)** - уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически значимо не отличается между тарифами "А" и "C"
* **Альтернативная гипотеза (H1)** - уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически значимо отличается между тарифами "А" и "C"

### Гипотеза № 3. Уровень оттока клиентов при использовании оптимального и неоптимального тарифного плана не отличается

Сформируем таблицу **df_proportion**, содержащую сведения об оттоке клиентов при использовании оптимального и неоптимального тарифного плана:

```python
df_proportion = pd.DataFrame(
    {
        'tariff_optimal_nonoptimal': ['optimal', 'nonoptimal'],
        'all_users': [((len(df_users_ch) - len(df_noneffective_churn)) + (len(df_users_ac) - len(df_noneffective))), 
                      (len(df_noneffective) + len(df_noneffective_churn))],
        'churn': [(len(df_users_ch) - len(df_noneffective_churn)), len(df_noneffective_churn)]
    }      
)
df_proportion['%churn'] = df_proportion['churn'] * 100 / df_proportion['all_users']
```

```python
df_proportion
```

Выдвинем гипотезы:
* **Нулевая гипотеза (H0)** - доля клиентов ушедших в отток при использовании оптимального тарифа не отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа
* **Альтернативная гипотеза (H1)** - доля клиентов ушедших в отток при использовании оптимального тарифа отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа

Проверку гипотез выполним методом двустороннего z-теста. Уровень значимости alpha=0,05.

```python
alpha = 0.05
count = np.array([df_proportion['churn'][0], df_proportion['churn'][1]])
nobs = np.array([df_proportion['all_users'][0], df_proportion['all_users'][1]])
stat, p_value = proportions_ztest(count, nobs,  alternative='two-sided')
if p_value > alpha:
    print('Нулевая гипотеза H0 не отвергнута: доля клиентов ушедших в отток при использовании', 
          'оптимального тарифа статистически значимо не отличается от доли клиентов ушедших',
          'в отток при использовании неоптимального тарифа')
else:
    print('Нулевая гипотеза H0 отвергнута: доля клиентов ушедших в отток при использовании', 
          'оптимального тарифа статистически значимо отличается от доли клиентов ушедших',
          'в отток при использовании неоптимального тарифа')
```

За ноябрь 2019 выручка провайдера "Нупозвони" составила 657938,1 рублей, из них дополнительный доход за счет использования клиентами неоптимальных тарифов - 187950,4 рублей, или 40%. Учитывая динамику ежемесячной переплаты клиентами, перевод всех клиентов провайдера на оптимальные тарифы приведет к потерям около 40% выручки.

Проверка выдвинутых гипотез показала следующее:

Проверка **Гипотезы № 1. Выручка провайдера не изменится при переводе клиентов на другие тарифы** показала, что при переводе всех клиентов на выгодные для них тарифы выручка провайдера статистически значимо изменится и уменьшится с 657938,1 рублей до 469987,7 рублей, что составляет 40%.

Проверка **Гипотезы № 2. Уровень дополнительного дохода за счет использования клиентами неоптимального тарифа не отличается для разных тарифных планов** показала, что уровень дополнительного дохода за счет использования клиентами неоптимального тарифа статистически значимо отличается для разных тарифов и составляет для тарифа "А" 3988 рублей, для тарифа "B" 992,6 рублей и для тарифа "C" 0 рублей.

Проверка **Гипотезы № 3. Уровень оттока клиентов при использовании оптимального и неоптимального тарифного плана не отличается** показала, что доля клиентов ушедших в отток при использовании оптимального тарифа (15,4%) статистически значимо не отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа (14,1%).

## Рекомендации

Учитывая, что доля клиентов ушедших в отток при использовании оптимального тарифа (15,4%) статистически значимо не отличается от доли клиентов ушедших в отток при использовании неоптимального тарифа (14,1%), а выручка провайдера существенно изменится - уменьшится на 40%, перевод клиентов на другие тарифы не является выгодным решением. Тем не менее, появление новых тарифов у конкурентов и возросший отток клиентов провайдера требует принятия мер. Решением проблемы удержания клиентов и снижения потерь в выручке может стать индивидуальный подход к клиентам в соответствии с их профилем. Изучение новых тарифов конкурентов и анализ их сильных и слабых мест позволит разработать успешную стратегию повышения лояльности клиентов провайдера.

В первую очередь необходимо обратить внимание на крупных клиентов, использующих тариф "А", поскольку выручка от них составляет треть всех доходов. Основные расходы такие клиенты несут на большое количество исходящих звонков и операторов. Снизить расходы клиентов на операторов позволит оптимизация их количества и отказ от услуг неэффективных операторов. В рамках стратегии можно уменьшить стоимость услуги каждого оператора при их определенном количестве, например, более 10, или предоставлять услуги оператора бесплатно при достижении определенного лимита, например, каждый 10 оператор бесплатно. Осуществить данные меры можно в рамках создания нового тарифа или временной акции. Это позволит не переводить клиента на другой тариф и при этом снизить его ежемесячные затраты, не влияя существенно на выручку провайдера. 

Стоит учитывать, что во второй месяц использования тарифа повышается риск оттока клиентов. Для клиентов с переплатой, превышающей средние значения в первый месяц, для тарифа "А" - 3988 рублей, тарифа "B" - 992,6 рублей возможно стоит предложить скидку от 10-20% на услуги провайдера. Это также позволит снизить затраты клиентов без существенного влияния на выручку провайдера. 

Перевести на другие тарифы необходимо тех клиентов, чьи выгодные тарифы оказались с более высокой базовой стоимостью, например, клиент использует тариф "С", более выгодным определен тариф "В", нужно предложить переход на "В".  В остальных случаях переход должен быть крайней мерой, возможно при попытке клиента отказаться от услуг провайдера.
